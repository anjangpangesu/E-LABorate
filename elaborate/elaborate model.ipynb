```python
# import packages
import numpy as np
import pandas as pd
import pickle

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.activations import linear, relu, sigmoid
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import optimizers, losses
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# import data
df = pd.read_csv('elaborate_new1.csv')
df #show data below
```

```python
df.drop(columns=['Unnamed: 0','Unnamed: 0.1'], inplace=True)
```

```python
df['Class'].unique()
```

```python
# change male and female to 0 and 1
df['Sex'][df['Sex'] == 'Male'] = 0
df['Sex'][df['Sex'] == 'Female'] = 1

#change class to 1,2,3,4
df['Class'][df['Class'] == 'Normal health status: no immediate need for medical consultation'] = 0
df['Class'][df['Class'] == 'Immediate medical consultation required'] = 3
df['Class'][df['Class'] == 'It is advisable to seek a medical consultation in order to facilitate a comprehensive assessment for further diagnostic purposes.'] = 1
df['Class'][df['Class'] == 'Medical consultation recommended for moderate condition'] = 2
df
```

```python
# split test and train data
X = df.drop(columns=['Interpretation', 'Age (year)', 'Class'],axis=1)
X = np.asarray(X).astype(np.float32)
X = StandardScaler().fit_transform(X)
y = df[['Class']]
y = np.asarray(y).astype(np.float32)
y = to_categorical(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
print(X_train.shape)
print(y_train.shape)

print(X_test.shape)
print(y_test.shape)
```

```python
#set callback
class myCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        if logs.get('val_accuracy') is not None and logs.get('val_accuracy') > 0.66 and logs.get('accuracy')>0.75:
            print("\nReached 66% validation accuracy so cancelling training!")
            self.model.stop_training = True
```

```python
#TRAINING DATA
callbacks=myCallback()

#set dense layer unit and input shape                             
model = Sequential([
    Dense(128, activation='relu', input_shape = X_train[0].shape, kernel_regularizer='l2'),
    tf.keras.layers.Dropout(0.5),
    Dense(64, activation='relu', kernel_regularizer='l2'),
    tf.keras.layers.Dropout(0.5),
    Dense(32, activation='relu', kernel_regularizer='l2'),
    Dense(4, activation='softmax')
])
model.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['accuracy'])
hist = model.fit(X_train, y_train, epochs=250,batch_size=5, callbacks=callbacks, validation_data=(X_test,y_test))
```

```python
loss, acc = model.evaluate(X_test, y_test)
print(f"Loss is {loss},\nAccuracy is {acc * 100}")
```

```python
import tensorflow as tf

# Convert the model
converter = tf.lite.TFLiteConverter.from_keras_model(model) # path to the SavedModel directory
tflite_model = converter.convert()
```

```python
# Save the model.
with open('elaborate_model.tflite', 'wb') as f:
  f.write(tflite_model)
```
